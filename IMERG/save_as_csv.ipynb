```python
import h5pyd as h5py
import pandas as pd
```

```python
# open file in read-only mode
f = h5py.File("/shared/NASA/IMERG/3B-MO.MS.MRG.3IMERG.20210701-S000000-E235959.07.V06B.HDF5", mode="r")
```

```python
# get the precipitation dataset
precip = f["/Grid/precipitation"]
precip
```

```python
# read every 100th element to numpy array
data = precip[0,::100, ::100]
data
```

```python
# convert numpy array to pandas dataframe
df = pd.DataFrame(data)
df
```

```python
# save dataframe to csv file
df.to_csv("mydata.csv")
```
